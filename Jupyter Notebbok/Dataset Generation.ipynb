```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 7000

# Generate random data
data = {
    'Age': np.random.randint(1, 90, n_samples),
    'Body Temperature': np.random.uniform(36.0, 40.0, n_samples),
    'Respiratory Rate': np.random.randint(12, 30, n_samples),
    'Heart Rate': np.random.randint(60, 120, n_samples),
    'Cough Severity': np.random.randint(1, 6, n_samples),
    'Chest Pain Severity': np.random.randint(1, 6, n_samples),
    'Oxygen Saturation': np.random.uniform(85.0, 100.0, n_samples),
    'Shortness of Breath': np.random.randint(0, 2, n_samples),  # 0 = No, 1 = Yes
    'Sputum Production': np.random.randint(0, 2, n_samples),    # 0 = No, 1 = Yes
    'History of Smoking': np.random.randint(0, 2, n_samples),   # 0 = No, 1 = Yes
    'History of Respiratory Infections': np.random.randint(0, 2, n_samples),  # 0 = No, 1 = Yes
    'Exposure to Pollutants': np.random.randint(0, 2, n_samples),  # 0 = No, 1 = Yes
    'Early Diagnosis': np.random.randint(0, 2, n_samples)  # 0 = No Disease, 1 = Acute Respiratory Disease
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
df.to_csv('respiratory.csv', sep = ',' , index=False, encoding='utf-8')
```
